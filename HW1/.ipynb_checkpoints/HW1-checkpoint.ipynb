```python
from scipy.io import loadmat as load
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
# import cupy as np
```

## Preparation
Prepare data from given MATLAB files.

```python
data = load('data.mat')
labels = load('label.mat')
train_data = data['imageTrain'].astype('int').reshape((784,5000)).T
train_labels = labels['labelTrain'].reshape((5000,))
test_data = data['imageTest'].astype('int').reshape((784,500)).T
test_labels = labels['labelTest'].reshape((500,))
```

## k nearest neighbor algorithm

```python
predictions = []
closest = []
for image in test_data:
    close = np.argmin(np.linalg.norm(image-train_data, axis=1))
    predictions.append(train_labels[close])
    closest.append(close)
```

## Problem 1
Use the Training images and the ground truth information to train a Nearest Neighbor Classifier
using Euclidean distance metric and test it on the test database.
Compute and plot the error rates for each class, i.e. calculate $$P(Error|Class = i) \quad \forall \quad i = 0 . . . 9 $$

```python
expected = []
predicted = [0] * 10
for i in range(10):
    expected.append(np.count_nonzero(test_labels == i))
for i in range(len(test_labels)):
    answer = test_labels[i]
    if answer == predictions[i]:
        predicted[answer] += 1
```

```python
plt.stem([1-predicted[i]/expected[i] for i in range(10)])
plt.xlabel('Class')
```

## Problem 2
Calculate the total error rate, i.e. compute $$P(Error)$$


```python
print(f'Total Error Rate: {1-sum(predicted)/sum(expected)}')
```

## Problem 3
For 5 test images which are misclassified by the NN Classifier, display both the test image and the
image from the training database which is closest to the test image. Can you explain, looking at
the test image and the corresponding nearest image, why did the NN classifier failed to perform?

```python
wrong = []
i = 0
while len(wrong) <= 4:
    if predictions[i] != test_labels[i]:
        wrong.append(i)
    i += 1
```

```python
for i in wrong:
    ax, arr = plt.subplots(1,2)
    arr[0].imshow(test_data[i].reshape((28,28)), cmap='gray')
    arr[0].set_yticklabels([])
    arr[0].set_xticklabels([])
    arr[0].set_title('Original Image')
    arr[0].set_xlabel(f'Original: {test_labels[i]}')
    arr[1].imshow(train_data[closest[i]].reshape((28,28)), cmap='gray')
    arr[1].set_title('Nearest Neighbor')
    arr[1].set_xlabel(f'Predicted: {train_labels[closest[i]]}')
    arr[1].set_yticklabels([])
    arr[1].set_xticklabels([])
```

    Despite the defined borders of the images, I can see that on incorrectly classified images, there is a lot of overlap with the nearest neighbor image. 
    For example, for the first pair of images, I can see that most of the pixels in the '4' is also included in the '9'. Hence, it is possible for when I elementwise subtract the pixels between the two images, more pixels would have lower values, which would directly reduce the 'total distance' between the two images. 
    Another example is the fourth pair of numbers, which is the absurdly drawn '4' and '6'. Every single pixel from the '4' looks like it is included in the '6'. Therefore, the norm of the difference would be very low. 
    Below is some code that visualizes the explanation. 

```python
test = [test_data[wrong[0]], test_data[wrong[4]]]
predictions = []
closest = []
for image in test:
    norms = np.linalg.norm(image-train_data, axis=1)
    ten = []
    for i in range(10):
        close = np.argmin(norms)
        predictions.append(train_labels[close])
        norms = np.delete(norms, close, 0)
        ten.append(close)
    closest.append(ten)
```

```python
best = []
for close in closest:
    for index in close:
        if train_labels[index] == 4:
            best.append(index)
            break
best
```

```python
ax, arr = plt.subplots(2,3)

arr[0][0].imshow(test[0].reshape((28,28)), cmap='gray')
arr[0][0].set_yticklabels([])
arr[0][0].set_xticklabels([])
arr[0][0].set_title('Original Image')
arr[0][0].set_xlabel(f'Original: {test_labels[wrong[0]]}')

arr[0][1].imshow(train_data[closest[0][0]].reshape((28,28)), cmap='gray')
arr[0][1].set_title('Nearest Neighbor')
arr[0][1].set_xlabel(f'Predicted: {train_labels[closest[0][0]]}')
arr[0][1].set_yticklabels([])
arr[0][1].set_xticklabels([])

arr[0][2].imshow(train_data[[best[0]]].reshape((28,28)), cmap='gray')
arr[0][2].set_title('Nearest \'4\'')
arr[0][2].set_xlabel(f'Class: {train_labels[best[0]]}')
arr[0][2].set_yticklabels([])
arr[0][2].set_xticklabels([])

image = test[0]-test[0]
arr[1][0].imshow(image.reshape((28,28)), cmap='gray')
arr[1][0].set_title('Original - Original', size=10)
arr[1][0].set_xlabel(f'Norm: {np.linalg.norm(image): .2f}')

image = test[0]-train_data[closest[0][0]]
arr[1][1].imshow(image.reshape((28,28)), cmap='gray')
arr[1][1].set_title('Original - Nearest', size=10)
arr[1][1].set_xlabel(f'Norm: {np.linalg.norm(image): .2f}')


image = test[0]-train_data[[best[0]]]
arr[1][2].imshow(image.reshape((28,28)), cmap='gray')
arr[1][2].set_title('Original - Nearest \'4\'', size=10)
arr[1][2].set_xlabel(f'Norm: {np.linalg.norm(image): .2f}')
```

We can see that there are training samples that have a less norm difference between the testing image and training image. Even less than the closest correct sample as well.
